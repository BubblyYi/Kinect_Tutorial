```python
def plot_jump_data(fname):
    with open(fname,'rb') as f:
        pos_li=pickle.load(f)
        h=[y[1] for y in pos_li]
        plt.plot(h)
        plt.show()
        dh=[y2-y1 for y1,y2 in zip(h[:-1],h[1:])]
        plt.plot(dh)
        plt.show()
```

```python
plot_jump_data('top_pos_list')
```
